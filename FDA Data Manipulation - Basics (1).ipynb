```r
a = c(rep("A",3),rep("B",3),rep("C",2))
b = c(1,1,2,4,1,1,2,2)
df= data.frame(a,b)
df
```

```r
duplicated(df)
```

```r
dfd1 = duplicated(df$a)
dfd1
class(dfd1)
dfd2 = duplicated(df$b)
dfd2
class(dfd2)
```

```r
dfu = unique(df)
dfu
```

```r
dfd = df[duplicated(df),]
dfd
```

```r
nrow(dfd)
nrow(dfu)
```
